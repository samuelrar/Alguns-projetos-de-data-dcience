# **INFORMAÇÕES SOBRE O DATASET**

Dados sobre o Programa Universidade para Todos (PROUNI) que concedeu bolsas de estudos em universidades particulares: **Cursos e notas de corte do PROUNI 2018** https://brasil.io/dataset/cursos-prouni/cursos/

Os dados em CSV estão disponíveis em https://drive.google.com/drive/u/0/folders/12E1nZP2iaitXf0sUJ6d8Qc9dWYU7mPgg

```python
import altair as alt
import pandas as pd
caminho = '/content/cursos-prouni.csv'

cursos = pd.read_csv(caminho)
```

```python
cursos.head()
```

# Informações sobre os cursos de Medicina ofertados no PROUNI

```python
medicina = cursos[cursos['nome'] == 'Medicina']
```

## Número de cursos por estado

```python
alt.Chart(medicina).mark_bar().encode(
    x = alt.X('uf_busca'),
    y = 'count()'
)
```

## Info sobre nota de corte e mensalidade para cada curso

O gráfico abaixo é interativo. Ao passar o mouse sobre cada ponto, serão dadas mais informações a respeito do curso.

```python
alt.Chart(medicina).mark_point().encode(

    x = alt.X(
        'nota_integral_ampla',
        scale = alt.Scale(zero = False),
        axis = alt.Axis(title = 'Nota de Corte em Ampla Concorrência')
        ),

    y = alt.Y(
        'mensalidade',
        scale = alt.Scale(zero = False),
        axis = alt.Axis(title = 'Mensalidade')
        ),

    tooltip = ['nota_integral_ampla', 'mensalidade', 'universidade_nome']

)
```

## Cursos por estado

Abaixo, um gráfico interativo que permite ao usuário a escolha de um estado brasileiro para obter informações detalhadas sobre os cursos de medicina ofertados no ProUni.

```python
selecao = alt.selection_multi(fields=['uf_busca'])

hist = alt.Chart(medicina).mark_bar().encode(

    x = alt.X(
        'uf_busca',
        axis = alt.Axis(title = 'Busca por UF')
        ),

    y = alt.Y(
        'count()',
        axis = alt.Axis(title = 'Número de cursos')
        ),

    color = alt.condition(selecao,'nome',alt.value('lightgray'), legend = None)
).add_selection(
    selecao
)

point = alt.Chart(medicina).mark_point().encode(
    x = alt.X('nota_integral_ampla',
              scale = alt.Scale(zero = False),
              axis = alt.Axis(title = 'Nota de Corte em Ampla Concorrência')
              ),

    y = alt.Y('mensalidade',
              scale = alt.Scale(zero = False),
              axis = alt.Axis(title = 'Mensalidade')
             ),

    tooltip = ['nota_integral_ampla', 'mensalidade', 'universidade_nome', 'cidade_busca']
).transform_filter(
    selecao
)

hist | point
```

# **Informações de cursos por busca**

O usuário deve fornecer um nome de curso disponível no dataset e será apresentado a ele um gráfico com as informações de busca.

```python
Curso = input(str('Busque aqui um curso para obter informações: '))

tabela_curso = cursos[cursos.nome == Curso]

selecao = alt.selection_multi(fields=['uf_busca'])

hist = alt.Chart(tabela_curso).mark_bar().encode(
    x = alt.X(
        'uf_busca',
        axis = alt.Axis(title = 'Busca por UF')
        ),

    y = alt.Y(
        'count()',
        axis = alt.Axis(title = 'Número de cursos')
        ),
    color = alt.condition(selecao,'nome',alt.value('lightgray'), legend = None)
).add_selection(
    selecao
)

point = alt.Chart(tabela_curso).mark_point().encode(
    x = alt.X('nota_integral_ampla',
              scale = alt.Scale(zero = False),
              axis = alt.Axis(title = 'Nota de Corte em Ampla Concorrência')
              ),

    y = alt.Y('mensalidade',
              scale = alt.Scale(zero = False),
              axis = alt.Axis(title = 'Mensalidade')
             ),
    tooltip = ['nota_integral_ampla', 'mensalidade', 'universidade_nome', 'cidade_busca', 'turno']
).transform_filter(
    selecao
).interactive()

hist | point
```

# Estudo sobre os cursos que o ProUNI oferta na cidade de Uberlândia - MG

```python
cursos_uberlandia = cursos[cursos['cidade_busca'] == 'Uberlandia']
```

## Lista de cursos disponíveis em ordem alfabética

```python
import numpy as np
cursos_udi = np.sort(cursos_uberlandia['nome'].unique())
print(cursos_udi)
```

```python
cursos_uberlandia['universidade_nome'].unique()
```

```python
selecao = alt.selection_multi(fields=['universidade_nome'])

hist = alt.Chart(cursos_uberlandia).mark_bar().encode(

    x = alt.X(
        'universidade_nome:N',
        axis = alt.Axis(title = 'Busca por Instituição de Ensino')
        ),

    y = alt.Y(
        'count()',
        axis = alt.Axis(title = 'Número de cursos')
        ),

    color = alt.condition(selecao,alt.value('steelblue'),alt.value('lightgray'), legend = None)
).add_selection(
    selecao
)

point = alt.Chart(cursos_uberlandia).mark_circle().encode(
    x = alt.X('nota_integral_ampla:Q',
              scale = alt.Scale(zero = False),
              axis = alt.Axis(title = 'Nota de Corte em Ampla Concorrência')
              ),

    y = alt.Y('mensalidade:Q',
              scale = alt.Scale(zero = False),
              axis = alt.Axis(title = 'Mensalidade')
             ),

    tooltip = ['nome', 'nota_integral_ampla', 'mensalidade']
).transform_filter(
    selecao
)

hist | point
```
